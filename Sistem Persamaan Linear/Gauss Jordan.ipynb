```python
import numpy as np
```

```python
def obe(matriks, hasil):
  for i in range(len(matriks)):
    if matriks[i][i] == 0:
      c = 0
      for j in range(i + 1, len(matriks)):
        if matriks[j][i] != 0:
          matriks[i], matriks[j] = matriks[j].copy(), matriks[i].copy()
          hasil[i], hasil[j] = hasil[j].copy(), hasil[i].copy()
          c = 1
          break
      if c == 0:
        hasil = "Matriks tidak memiliki solusi cikðŸ˜‹"
        return matriks, hasil
    a = matriks[i][i]
    matriks[i][i] = matriks[i][i] * (1/a)
    hasil[i] = hasil[i] * (1/ a)
    for j in range(len(matriks[i])):
      if i != j:
        matriks[i][j] /= a
    for k in range(len(matriks)):
      if i != k:
        pembuatNol = matriks[k][i]
        for l in range(len(matriks[k])):
          matriks[k][l] -= pembuatNol * matriks[i][l]
        hasil[k] -= pembuatNol * hasil[i]
  return matriks, hasil
```

##Soal No 1

```python
matriks = np.array([[4, -3, 1], [2, 1, -4], [1, 2, -2]]).astype(float)
hasil = np.array([11, -1, 1]).astype(float)

matriks, hasil = obe(matriks, hasil)
print(f"matriks: \n{matriks}")
print(f"hasil: \n{hasil}")
```
