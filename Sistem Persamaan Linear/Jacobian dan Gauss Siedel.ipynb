```python
import numpy as np
import pandas as pd
import sympy as sym
```

#Soal No 1

Metode Jacobi

```python
x, y, z = sym.symbols('x y z')
pers = [sym.Eq(4*x - 3*y + z, 11), sym.Eq(2*x + y - 4*z, -1), sym.Eq(x + 2*y - 2*z, 1)]
persX = sym.solve(pers[0], x)
persY = sym.solve(pers[1], y)
persZ = sym.solve(pers[2], z)
X = 0
Y = 0
Z = 0
hasil = None
iterations_data = []

tolerance = 10**-4
for i in range(50):
    XSementara = float(X)
    YSementara = float(Y)
    ZSementara = float(Z)

    X = persX[0].subs({y: YSementara, z: ZSementara})
    Y = persY[0].subs({x: XSementara, z: ZSementara})
    Z = persZ[0].subs({x: XSementara, y: YSementara})

    error = np.array([[abs((X-XSementara)/X)/100], [abs((Y-YSementara)/Y)/100], [abs((Z-ZSementara)/Z)/100]])

    iterations_data.append([i+1, float(X), float(Y), float(Z), float(error[0]), float(error[1]), float(error[2])])

    if all(e < tolerance for e in error):
        hasil = X, Y, Z
        break
table = pd.DataFrame(iterations_data, columns=['Iteration', 'X', 'Y', 'Z', 'X Error', 'Y Error', 'Z Error'])
print(table)
if hasil is not None:
    print(hasil)
else:
    print('\nSolusi tidak ditemukan cik')
```

#Soal No 2

Metode Jacobi

```python
x1, x2, x3, x4 = sym.symbols('x1 x2 x3 x4')
pers = [sym.Eq(1*x1 - 1*x2 + 1*x3 - 2*x4, 8), sym.Eq(2*x1 - 1*x2 + 2*x3 - 1*x4, 5),
        sym.Eq(-1*x1 + 1*x2 + 2*x3 - 1*x4, 4), sym.Eq(1*x1 + 2*x2 + 4*x3 + 1*x4, 5)]
pers1 = sym.solve(pers[0], x1)
pers2 = sym.solve(pers[1], x2)
pers3 = sym.solve(pers[2], x3)
pers4 = sym.solve(pers[3], x4)
X1, X2, X3, X4 = 0, 0, 0, 0
hasil = None
iterations_data = []

for i in range(50):
    X1Sementara, X2Sementara, X3Sementara, X4Sementara = X1, X2, X3, X4
    X1 = pers1[0].subs({x2: X2Sementara, x3: X3Sementara, x4: X4Sementara})
    X2 = pers2[0].subs({x1: X1Sementara, x3: X3Sementara, x4: X4Sementara})
    X3 = pers3[0].subs({x1: X1Sementara, x2: X2Sementara, x4: X4Sementara})
    X4 = pers4[0].subs({x1: X1Sementara, x2: X2Sementara, x3: X3Sementara})
    error = [abs((X1 - X1Sementara) / X1) / 100, abs((X2 - X2Sementara) / X2) / 100, abs((X3 - X3Sementara) / X3) / 100, abs((X4 - X4Sementara) / X4) / 100]
    iterations_data.append([i + 1, float(X1), float(X2), float(X3), float(X4),float(error[0]), float(error[1]), float(error[2]), float(error[3])])
    if all(e < tolerance for e in error):
        hasil = X1, X2, X3, X4
        break
columns = ['Iteration', 'X1', 'X2', 'X3', 'X4', 'X1 Error', 'X2 Error', 'X3 Error', 'X4 Error']
table = pd.DataFrame(iterations_data, columns=columns)
print(table)
if hasil is not None:
    print(hasil)
else:
    print('\nSolusi tidak ditemukan cik')
```
