# sin x

## Ketika x = 1

```python
x = 1
n = 1
s = 0
galat = 1
while galat > 10**(-6):
    s += ((-1)**(n-1)) * (x**(2*n-1)) / math.factorial(2*n-1)
    galat = abs(s - math.sin(x))
    n += 1
print(f"Jumlah suku yang dijumlahkan: {n-1}")
print(f"Nilai sebenarnya sin({x}): {math.sin(x)}")
print(f"Hasil perhitungan: {s}")
print(f"Galat: {galat}")
```

#Log x

## Ketika x = 1

```python
x = 1
n = 1
s = 0
galat = 1
while galat > 10**(-6):
  suku = (1 - x)**n / n
  s -= suku
  galat = abs(s - math.log(1 - (1 - x)))
  n += 1
print(f"Jumlah suku yang dijumlahkan: {n-1}")
print(f"Nilai sebenarnya ln(1 - (1 - {x})): {math.log(1 - (1 - x))}")
print(f"Hasil perhitungan: {s}")
print(f"Galat: {galat}")
```
