```python
import math
import pandas as pd
```

#Soal 1

## Metode Regulasi Falsi

```python
def f(x):
    return (x**3) - (7*x) + 1
a = 1
b = 3
n = 40
data = []

if f(a) * f(b) < 0:
    for i in range(n):
        c = ((f(b) * a) - (f(a) * b)) / (f(b) - f(a))
        data.append([i + 1, a, b, c, f(c)])

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

        if abs(f(c)) < 10**-3:
            hasil = c
            break
else:
    print("No root")

df = pd.DataFrame(data, columns=['Iterasi', 'a', 'b', 'c', 'f(c)'])
print(df)
print(f"Nilai X = {hasil}")
```

## Metode Newton

```python
def f(x):
    return (x**3) - (7*x) + 1

def fTurunan(x):
    return (3*(x**2)) - 7

x = 0
n = 40
data = []

for i in range(n):
    x = x - f(x) / fTurunan(x)
    data.append([i+1, x])

    if f(x) < 0.001:
      hasil = x
      break

df = pd.DataFrame(data, columns=['Iterasi', 'Nilai X'])

print(df)
print(f"Nilai X = {hasil}")
```

## Metode Secant

```python
def f(x):
    return x**3 - 7*x + 1

x0 = 1
x1 = 3
n = 40
data = []

for i in range(n):
    x = x1 - (f(x1) * (x0 - x1)) / (f(x0) - f(x1))
    data.append([i+1, x])

    if abs(f(x)) < 0.001:
        hasil = x
        break
    x0, x1 = x1, x

df = pd.DataFrame(data, columns=['Iterasi', 'Nilai X'])

print(df)
print(f"Nilai X = {hasil}")
```
