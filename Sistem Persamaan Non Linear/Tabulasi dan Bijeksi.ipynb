#Soal No 1

## Menggunakan Metode Tabel

```python
xBawah = 0
xAtas = 5
jumlahPembagi = 40
step = (xAtas - xBawah) / jumlahPembagi
i = 0
data = []
while (i <= jumlahPembagi and xBawah < xAtas):
  x=xBawah+(i*step)
  fx1 = (x**3-7*x+1)
  fx2 = ((x+1)**3-7*(x+1)+1)
  if(fx1==0):
    hasil = x
  elif(fx1*fx2<0):
    hasil = x
    if abs(fx1)<abs(fx2):
      hasil = f"Hasilnya hampir sama dengan: {x}"
    else:
      hasil = f"Hasilnya diantara {x} dan {x+1}"
  data.append([i, round(x, 2), round(fx1, 7)])
  i = i + 1
if xBawah > xAtas:
  print("batas atas harus lebih kecil dari batas bawah")
df = pd.DataFrame(data, columns=['i', 'x', 'fx1'])
pd.set_option('display.max_rows', None)
print(df)
```
